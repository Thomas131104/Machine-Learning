## **I. Các thao tác cơ bản trên Sklearn**

- Sử dụng module: **`LinearRegression`** trong **`sklearn.linear_model`**

- Khởi tạo mô hình: **`model = LinearRegression()`**

- Huấn luyện mô hình: **`model.fit(X, y)`**

- Dự đoán: **`y_pred = model.predict(X_test)`**


## **II. Đánh giá mô hình**

### **1. Mean Absolute Error (MAE):**
- Đo lường trung bình độ lỗi tuyệt đối giữa dự đoán và giá trị thực tế.
- Càng thấp, mô hình càng tốt.
- Sử dụng:
```
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true, y_pred)

```


### **2. Mean Squared Error** (MSE)
- Đo lường trung bình của bình phương của độ lỗi giữa dự đoán và giá trị thực tế.
- Càng thấp, mô hình càng tốt.
- Sử dụng
```
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true, y_pred)
```


### **3. Root Mean Squared Error** (RMSE)
- Căn bậc hai của MSE, cho biết độ lớn của sai số trung bình.
- Càng thấp, mô hình càng tốt.
- Sử dụng
```
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
```


### **4. Hệ số R**
- Đo lường tỷ lệ phương sai của biến mục tiêu mà mô hình có thể giải thích.
- Sử dụng
```
from sklearn.metrics import r2_score
r2 = r2_score(y_true, y_pred)

```

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
````

````python
# Phân chia dữ liệu

X = df[['CC', 'GK', 'CK']]
y = df['TK']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)


# Xây dựng mô hình LR
LR = LinearRegression()
LR.fit(X, y)


# Tính toán hệ số tương quan
y_pred = LR.predict(X_test)
r = r2_score(y_test, y_pred)
print('Hệ số tương quan: ', r)


# Tính toán hệ số MSE
mse = mean_squared_error(y_test, y_pred)
print('MSE: ', mse)


# Viết phương trình hồi quy
print('Hệ số hồi quy:', LR.coef_)
print('Intercept:', LR.intercept_)
````
