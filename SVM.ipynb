# **SVM - Support Vector Machine**

## **I. Khái niệm**

- SVM là một thuật toán học máy phổ biến được sử dụng cho cả bài toán phân loại và hồi quy. SVM hoạt động bằng cách tạo ra một ranh giới phân chia (decision boundary) giữa các lớp dữ liệu sao cho khoảng cách từ các điểm dữ liệu gần nhất tới ranh giới này (gọi là các vector hỗ trợ) là lớn nhất có thể. Thuật toán SVM cũng có thể được mở rộng để xử lý các vấn đề phân loại phi tuyến tính bằng cách sử dụng các hàm nhân (kernel functions) để ánh xạ dữ liệu vào không gian cao chiều hơn.

# **II. Cách hoạt động**

- Tìm ranh giới phân chia: SVM cố gắng tìm một ranh giới phân chia tối ưu nhất (hyperplane) để phân tách các điểm dữ liệu của các lớp khác nhau. Hyperplane là một mặt phẳng (trong không gian hai chiều), hoặc một siêu mặt phẳng (trong không gian nhiều chiều), mà phân tách các điểm dữ liệu.

- Hỗ trợ vector: SVM xác định các điểm dữ liệu quan trọng nhất gần với ranh giới phân chia này, gọi là các vector hỗ trợ (support vectors). Các vector hỗ trợ là những điểm dữ liệu đặc biệt vì chúng nằm gần nhất với ranh giới phân chia và có vai trò quan trọng trong việc xác định hyperplane.

- Tối ưu hóa khoảng cách: SVM tối ưu hóa khoảng cách (margin) giữa các vector hỗ trợ và hyperplane. Mục tiêu là tìm ra hyperplane sao cho khoảng cách từ các vector hỗ trợ tới hyperplane là lớn nhất có thể. Điều này giúp SVM có khả năng tổng quát hóa tốt hơn khi đối mặt với dữ liệu mới.

- Xử lý dữ liệu phi tuyến tính: Ngoài ra, SVM có thể sử dụng các hàm nhân (kernel functions) để ánh xạ dữ liệu vào không gian cao chiều hơn, giúp SVM xử lý các bài toán phân loại phi tuyến tính. Các hàm nhân cho phép SVM tạo ra các ranh giới phân chia phức tạp hơn mà không cần phải biến đổi dữ liệu ban đầu một cách rõ ràng.

# **III. Ưu điểm**

- Hiệu suất tốt trong không gian chiều cao: SVM hiệu quả khi số chiều dữ liệu lớn hơn số lượng điểm dữ liệu, tức là trong không gian chiều cao.

- Hiệu quả khi số lượng điểm dữ liệu lớn: SVM vẫn hoạt động hiệu quả khi có một lượng lớn các điểm dữ liệu.

- Khả năng xử lý không gian dữ liệu phi tuyến tính: SVM có thể sử dụng kernel để phân loại dữ liệu phi tuyến tính, giúp nó mở rộng ứng dụng cho nhiều loại dữ liệu phức tạp hơn.

- Hạn chế overfitting: SVM có các tham số, như tham số regularization (C) và sự lựa chọn kernel, giúp kiểm soát và giảm thiểu hiện tượng overfitting.

- Hỗ trợ vectors quan trọng: SVM tìm ra các vectors quan trọng (support vectors) quyết định vị trí của đường phân chia, làm giảm chi phí tính toán khi áp dụng mô hình cho dữ liệu mới.

# **IV. Nhược điểm**

- Khó khăn trong việc chọn kernel: Việc chọn kernel phù hợp có thể khó khăn và phụ thuộc vào tính chất của dữ liệu. Một lựa chọn kernel không tốt có thể dẫn đến kết quả phân loại không chính xác.

- Tốc độ huấn luyện chậm với số lượng dữ liệu lớn: SVM có thể yêu cầu nhiều thời gian để huấn luyện khi số lượng điểm dữ liệu lớn, đặc biệt là khi sử dụng kernel phức tạp.

- Khó khăn trong xử lý dữ liệu có nhiễu lớn: SVM dễ bị ảnh hưởng bởi dữ liệu nhiễu, đặc biệt là khi biên phân chia giữa các lớp dữ liệu không rõ ràng.

- Khó khăn trong việc diễn giải kết quả: Đôi khi đường phân chia của SVM không dễ diễn giải hoặc không trực quan đối với người dùng, đặc biệt đối với các kernel phức tạp.

- Khó khăn trong việc xử lý dữ liệu lớn: SVM yêu cầu bộ nhớ và tài nguyên tính toán lớn hơn so với một số thuật toán học máy khác khi xử lý dữ liệu lớn.

# V. **Các hàm phổ biến trong sklearn**

**1. Module `sklearn.svm`**

- **SVC**: Đây là lớp chính trong scikit-learn để huấn luyện mô hình SVM cho bài toán phân loại. Hàm này hỗ trợ nhiều loại kernel và có các tham số quan trọng như kernel, C (tham số regularization), gamma (cho các kernel như RBF),...

- **SVR**: Đây là lớp để huấn luyện mô hình SVM cho bài toán hồi quy. Tương tự như SVC, SVR cũng hỗ trợ nhiều loại kernel.

- **NuSVC** và **NuSVR**: Là các biến thể của SVC và SVR với hỗ trợ cho các tham số nu thay vì C. nu là một tham số khác để điều chỉnh sự đơn giản hóa (regularization) trong SVM.

- **OneClassSVM**: Được sử dụng cho bài toán phát hiện ngoại lệ (outlier detection) và học không giám sát (unsupervised learning).

- **LinearSVC**: Là một lớp SVM dựa trên liblinear, được sử dụng cho bài toán phân loại nhị phân với kernel tuyến tính.



**2. Các phương thức và thuộc tính**

**a. Các phương thức chung**

- decision_function(X): Tính toán giá trị của hàm quyết định cho dữ liệu X.

- fit(X, y): Huấn luyện mô hình với dữ liệu X và nhãn y.

- predict(X): Dự đoán nhãn hoặc giá trị đầu ra cho dữ liệu X.

- score(X, y): Đánh giá hiệu suất của mô hình trên dữ liệu X với nhãn y.

- support_vectors_: Các vector hỗ trợ (support vectors) của mô hình.

- n_support_: Số lượng vector hỗ trợ cho từng lớp.

- coef_: Các hệ số của đường phân chia hoặc mô hình hồi quy.



**b. Trong `sklearn.svm.SCV`**

* Thuộc tính:

- support_vectors_: Các vector hỗ trợ (support vectors).

- dual_coef_: Các hệ số Lagrange nhân với vector hỗ trợ để tính toán hàm quyết định.

- coef_: Các hệ số của đường phân chia trong trường hợp kernel tuyến tính.

- intercept_: Điểm cắt (bias term) của đường phân chia.


* Phương thức:

- fit(X, y): Huấn luyện mô hình SVM với dữ liệu X và nhãn y.

- predict(X): Dự đoán nhãn cho dữ liệu X.

- decision_function(X): Tính toán giá trị của hàm quyết định cho dữ liệu X.

- score(X, y): Đánh giá hiệu suất của mô hình trên dữ liệu X với nhãn y.



**c. Trong `sklearn.svm.SVR`**

* Thuộc tính:

- support_: Chỉ số của các vector hỗ trợ.

- dual_coef_: Các hệ số Lagrange nhân với vector hỗ trợ.

- coef_: Các hệ số của mô hình hồi quy trong trường hợp kernel tuyến tính.

- intercept_: Điểm cắt (bias term) của mô hình hồi quy.


* Phương thức:

- fit(X, y): Huấn luyện mô hình SVM hồi quy với dữ liệu X và giá trị y.

- predict(X): Dự đoán giá trị đầu ra cho dữ liệu X.

- score(X, y): Đánh giá hiệu suất của mô hình hồi quy trên dữ liệu X với giá trị y.

````python
import numpy as np
import random

size = 100
data = np.arange(0, 10.2, 0.2)

random.seed(100)
x1 = np.array(random.choices(data, k = size))
x2 = np.array(random.choices(data, k = size))

(x1, x2)
````

````python
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu và nhãn
# Ví dụ: Phân loại dữ liệu thành hai lớp dựa trên đường thẳng x1 = x2
y = np.where(x1 >= x2, 0, 1)  # Giả sử x1 >= x2 là lớp 0, ngược lại là lớp 1

# Huấn luyện mô hình SVM
svm_model = SVC(kernel='linear')
svm_model.fit(np.vstack((x1, x2)).T, y)

# Visualize đường phân chia
plt.figure(figsize=(8, 6))
plt.scatter(x1[y==0], x2[y==0], color='blue', label='Class 0')
plt.scatter(x1[y==1], x2[y==1], color='red', label='Class 1')

# Vẽ đường phân chia
w = svm_model.coef_[0]
b = svm_model.intercept_[0]
slope = -w[0] / w[1]
intercept = -b / w[1]
x_plot = np.linspace(0, 10, 100)
y_plot = slope * x_plot + intercept
plt.plot(x_plot, y_plot, color='green', linestyle='--', label='Decision Boundary')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('SVM Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
````

````python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu
size = 100
data = np.arange(0, 10.2, 0.2)
np.random.seed(100)
x1 = np.random.choice(data, size=size)
x2 = np.random.choice(data, size=size)

# Xác định nhãn dựa trên ví dụ đơn giản
y = np.where(x1 >= x2, 0, 1)  # Ví dụ: nhãn 0 khi x1 >= x2, ngược lại là nhãn 1

# Huấn luyện mô hình SVM
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(np.column_stack((x1, x2)), y)

# Vẽ đồ thị để minh họa ranh giới phân chia
plt.figure(figsize=(8, 6))
plt.scatter(x1[y==0], x2[y==0], color='blue', label='Class 0')
plt.scatter(x1[y==1], x2[y==1], color='red', label='Class 1')

# Vẽ ranh giới phân chia
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Tạo lưới điểm để vẽ ranh giới phân chia
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))
Z = svm_model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Vẽ đường phân chia và margin
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
            linestyles=['--', '-', '--'])
plt.title('SVM Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.show()
````
